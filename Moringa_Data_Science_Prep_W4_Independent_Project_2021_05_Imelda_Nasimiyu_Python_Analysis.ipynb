AUTOLIB DATASET ANALYSIS

```python
#importing required libraries
import pandas as pd
import numpy as np
```

```python
#importing required dataset for AutoLib

Autolib_DF = pd.read_csv("Autolib_dataset.csv")
Autolib_DF.head(10)
```

```python
Autolib_DF.isnull()
```

```python
#Merging hour and minute to one column
Autolib_DF['time'] = (pd.to_datetime(Autolib_DF['hour'].astype(str) + ':' + Autolib_DF['minute'].astype(str), format='%H:%M')
          .dt.time)
Autolib_DF.head()

#Merging year, month and day into one column
Autolib_DF['date'] = (pd.to_datetime(Autolib_DF['year'].astype(str) + "/" + Autolib_DF["month"].astype(str) + "/" + Autolib_DF["day"].astype(str)))
Autolib_DF.head()
```

```python
#Dropping the extra columns
Autolib_DF = Autolib_DF.drop(['year','month','day','hour','minute'],axis=1).head()
Autolib_DF.head()
```

```python
#Updating the column['Cars'] to include sum of all 3 types of cars
Autolib_DF.Cars = Autolib_DF['Bluecar counter'] +  Autolib_DF['Utilib counter'] + Autolib_DF['Utilib 1.4 counter']   
Autolib_DF.head()
```

```python
#Updating dataframe to include usage computation to determine whether the car was picked up, returned, or nothing happened.
#Car returned
Autolib_DF['Returned'] = Autolib_DF['Cars'] - Autolib_DF['Slots']
Autolib_DF.head(10)
```

```python
#Car in use / Also equal to slots
Autolib_DF['operational'] = Autolib_DF['Cars'] - Autolib_DF['Returned']
Autolib_DF.head(10)
```
